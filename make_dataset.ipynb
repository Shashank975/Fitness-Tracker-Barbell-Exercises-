```python
# Turn all this process into a function (You can remove the above files)
```

```python
files =glob("MetaMotion/MetaMotion/*.csv")

def read_data_from_files(files):
    acc_df = pd.DataFrame()
    gyr_df = pd.DataFrame()
    
    acc_set = 1
    gyr_set = 1
    
    
    # Initialize empty DataFrames

    for f in files:
        participant = f.split('\\')[1][0]
        label = f.split("-")[1]
        category = f.split("-")[2].rstrip("123").rstrip("_MetaWear_2019")

        df = pd.read_csv(f)

        df["participant"] = participant
        df["label"] = label
        df["category"] = category

        if "Accelerometer" in f:
            df["set"] = acc_set
            acc_set += 1
            acc_df = pd.concat([acc_df, df], ignore_index=True)
        if "Gyroscope" in f:
            df["set"] = gyr_set
            gyr_set += 1
            gyr_df = pd.concat([gyr_df, df], ignore_index=True)
    
    # Convert epoch (ms) to datetime and set as index for acc_df
    acc_df.index = pd.to_datetime(acc_df["epoch (ms)"], unit="ms")

    # Convert epoch (ms) to datetime and set as index for gyr_df
    gyr_df.index = pd.to_datetime(gyr_df["epoch (ms)"], unit="ms")

    del acc_df["epoch (ms)"]
    del acc_df["time (01:00)"]
    del acc_df["elapsed (s)"]
    del gyr_df["epoch (ms)"]
    del gyr_df["time (01:00)"]
    del gyr_df["elapsed (s)"]
    
    return acc_df,gyr_df
```

```python
acc_df,gyr_df = read_data_from_files(files)
```

```python
# Merging the files
```

```python
data_merged= pd.concat([acc_df.iloc[:,:3],gyr_df],axis = 1)
data_merged
```

```python
data_merged.columns = [
    "acc_x",
    "acc_y",
    "acc_z",
    "gyr_x",
    "gyr_y",
    "gyr_z",
    "participant",
    "label",
    "category",
    "set",
]
```

```python
# Resample the Data
# Accelerometer  : 12.500Hz (every point 0.88 sec)
# Gyroscope      : 25.000Hz (every point 0.4 sec)
```

```python
sampling = {
    "acc_x" : "mean",
    "acc_y" : "mean",
    "acc_z" : "mean",
    "gyr_x" : "mean",
    "gyr_y" : "mean",
    "gyr_z" : "mean",
    "participant" : "last",
    "label": "last",
    "category": "last",
    "set": "last",
}
```

```python
data_merged[:1000].resample(rule="200ms").apply(sampling)
```

```python
# Split by Days
```

```python
days = [g for n, g in data_merged.groupby(pd.Grouper(freq="D"))]
```

```python
data_resampled = pd.concat([df.resample(rule = "200ms").apply(sampling).dropna() for df in days])
data_resampled
```
